# Bank Marketing Data Analysis
This notebook presents an analysis of the `bankmarketing.csv` dataset, which includes data related to a bank's marketing campaigns. The main goal is to understand customer behavior and predict whether a client will subscribe to a term deposit.

```python
import pandas as pd

# Load the dataset - done
df = pd.read_csv('bankmarketing.csv')

# Display the first few rows - done
df.head()
```

```python
# Check for missing values and data types - done
df.info()

# Count of missing values in each column - additional step
df.isnull().sum()
```

```python
# Summary statistics for numerical columns - segregated
df.describe()
```

```python
# Summary statistics for categorical columns - segregated
df.describe(include='object')
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Set style
sns.set(style="whitegrid")

# Plot the distribution of the target variable - done
plt.figure(figsize=(6,4))
sns.countplot(x='y', data=df)
plt.title('Distribution of Term Deposit Subscription')
plt.xlabel('Subscribed to Term Deposit')
plt.ylabel('Count')
plt.show()

# Also show percentage distribution
print(df['y'].value_counts(normalize=True) * 100)
```

```python
# Correlation matrix for numerical features - done
# Select only numerical features
numeric_df = df.select_dtypes(include=['int64', 'float64'])

# Compute correlation matrix
corr = numeric_df.corr()

# Plot correlation heatmap
plt.figure(figsize=(10,6))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f", linewidths=0.5)
plt.title('Correlation Matrix for Numerical Features')
plt.show()
```
